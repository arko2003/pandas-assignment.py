```python
# f.replicating the split-apply-combine without groupby
averages={}
#spilting the data into different regions
for make in data['make'].unique():
    tempdf= data[data['make']==make]
#apply an aggregration fuunction 
    average= tempdf['price'].mean()
#combine the data into dataframe
    averages[make]= [average]
aggregate_data = pd.DataFrame.from_dict(averages, orient= 'index', columns= ['price'])
aggregate_data


```
